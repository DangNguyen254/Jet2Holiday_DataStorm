```python
import pandas as pd
import os
print("Current working directory:", os.getcwd())
print("\nFiles in current directory:", os.listdir('.'))
```

```python
parent_dir = os.path.dirname(os.getcwd())
train_df = pd.read_parquet(os.path.join(parent_dir, "data/processed/train_features.parquet"))
eval_df = pd.read_parquet(os.path.join(parent_dir, "data/processed/eval_features.parquet"))
train_cols = train_df.columns.tolist()
eval_cols = eval_df.columns.tolist()
print("train data columns: " + str(train_cols))
print("eval data columns: " + str(eval_cols))
for col in train_cols:
    if col not in eval_cols:
        print("column " + col + " is missing in eval data")
```
